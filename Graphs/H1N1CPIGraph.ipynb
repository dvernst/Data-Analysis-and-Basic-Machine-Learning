```python
labels = ['AI', 'C&BP', 'F&E', 'D&R', 'F&V','Hus','Apl','Tsp','MC','Rct','E&C','OF']
JanToFeb = [.4,-.5,-.1,-2.4,.4,0,1.3,1.9 ,.3,.4,.2,2]
FebToMar = [-.1,-.2,-.9,-2.4,.2,-.1,-.2,-1.1,.2,0, .2,2.7]
MarToApr = [0,-.7,0,-1.3,0,-.1 ,-.2 ,-.4 ,.4 ,-.4 ,.3 ,2.6 ]

x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, JanToFeb, width, label='JanToFeb')
rects2 = ax.bar(x + width/2, FebToMar, width, label='FebToMar')
rects3 = ax.bar(x + width*1.5, MarToApr, width, label='MarToApr')


ax.set_ylabel('Precensent of Change')
ax.set_title('H1NI CPI for All Urban Consumers ')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.grid(zorder=0)

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


#autolabel(rects1)
#autolabel(rects2)
#autolabel(rects3)

fig.tight_layout()

plt.show()
```
