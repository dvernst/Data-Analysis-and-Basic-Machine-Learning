```python
import pandas as pd
import numpy as np
```

```python
dfAPR = pd.read_csv('../Initial_Data_Files/CPI_SanFran_APR20.csv')
dfDEC = pd.read_csv('../Initial_Data_Files/CPI_SanFran_DEC19.csv')
```

```python
dfAPR
dfDEC
```

```python
dfAPR = dfAPR.iloc[:, [0, 1, 3]]
dfDEC = dfDEC.iloc[:, [0, 1, 3]]
```

```python
dfAPR.columns = ['Item and Group', 'IDX-FEB-2020', 'IDX-APR-2020']
```

```python
dfAPR.head(20)
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
dfAPR = dfAPR.iloc[[4, 6, 9, 10, 11, 12, 20, 33, 34, 45, 46, 47, 49]]
```

```python
dfDEC = dfDEC.iloc[[4, 6, 9, 10, 11, 12, 19, 31, 32, 43, 44, 45, 47]]
```

```python
dfAPR = dfAPR.reset_index(drop = True)
dfDEC = dfDEC.reset_index(drop = True)
```

```python
df = dfAPR
df.insert(1, 'IDX-OCT-2019', dfDEC.iloc[:, [1]])
df.insert(2, 'IDX-DEC-2019', dfDEC.iloc[:, [2]])
```

```python
df
```

```python
df = df.astype({'Item and Group': str, 'IDX-OCT-2019': 'float64', 'IDX-DEC-2019': 'float64', 'IDX-FEB-2020': 'float64', 'IDX-APR-2020': 'float64'})
df.dtypes
```

```python
df['Item and Group'] = df['Item and Group'].str.strip('.123456789() ')
```

```python
df['Inflation Change FEB->APR'] = ((df['IDX-APR-2020'] - df['IDX-FEB-2020']) / df['IDX-FEB-2020']) * 100
```

```python
FEB_20_SUM = df.loc[[1, 6, 7, 8, 9, 10, 11, 12], 'IDX-FEB-2020'].sum()
APR_20_SUM = df.loc[[1, 6, 7, 8, 9, 10, 11, 12], 'IDX-APR-2020'].sum()

TOTAL_INFL_PERC = ((APR_20_SUM - FEB_20_SUM) / FEB_20_SUM) * 100

TOTAL_INFL_PERC
```
