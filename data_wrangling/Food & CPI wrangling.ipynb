```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("CPI-Export-Import.csv")
df2 = pd.read_csv("H1N1_food_data.csv")
df2
```

```python
newFood = df2.loc[:,['breakfast', 'meat', 'dairy', 'fruits']]
newCost_living = df2.loc[:, ['Housing', 'apparel', 'transport', 'medical', 'recreation', 'education', 'others']]
newCost_living
```

```python
newFood["food_total"] = newFood.sum(axis=1)
newCost_living["cost_of_living"] = newCost_living.sum(axis=1)
print(newFood)
print(newCost_living)
```

```python
newdf2 = pd.concat([newFood, newCost_living], axis=1)
newdf2
```

```python
df = pd.concat([df1, newdf2], axis=1)
df
```

```python
df.to_csv('./CPI_Food_H1N1_updated.csv', index = False)
```
