```python
import pandas as pd
import numpy as np
```

```python
dfAPR = pd.read_csv('../Initial_Data_Files/CPI_LosAngeles_APR20.csv')
dfFEB = pd.read_csv('../Initial_Data_Files/CPI_LosAngeles_FEB20.csv')
```

```python
df = dfAPR.iloc[:, [0, 1, 2, 3]]
dfFEB = dfFEB.iloc[:, [0, 1, 2, 3]]
```

```python
df.columns = ['Item and Group', 'IDX-FEB-2020', 'IDX-MAR-2020', 'IDX-APR-2020']
```

```python
df.insert(1, 'IDX-DEC-2019', dfFEB.iloc[:, [1]])
df.insert(2, 'IDX-JAN-2020', dfFEB.iloc[:, [2]])
```

```python
df
```

### ONLY run next line ONCE

```python
df = df.drop([0, 1, 2, 3])
```

```python
df = df.reset_index(drop = True)
```

```python
df = df.astype({'Item and Group': str, 'IDX-DEC-2019': 'float64', 'IDX-JAN-2020': 'float64', 'IDX-FEB-2020': 'float64', 'IDX-MAR-2020': 'float64', 'IDX-APR-2020': 'float64'})
df.dtypes
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
df
```

```python
df = df.iloc[[0, 2, 5, 6, 7, 8, 16, 29, 30, 41, 42, 43, 45]]
df = df.reset_index(drop = True)
```

```python
df['Item and Group'] = df['Item and Group'].str.strip('.123456789() ')
```

```python
df['Inflation Change JAN->MAR'] = ((df['IDX-MAR-2020'] - df['IDX-JAN-2020']) / df['IDX-JAN-2020']) * 100
```

```python
df
```

```python
JAN_20_SUM = df.loc[[1, 6, 7, 8, 9, 10 , 11, 12], 'IDX-JAN-2020'].sum()
MAR_20_SUM = df.loc[[1, 6, 7, 8, 9, 10 , 11, 12], 'IDX-MAR-2020'].sum()

TOTAL_INFL_PERC = ((MAR_20_SUM - JAN_20_SUM) / JAN_20_SUM) * 100

TOTAL_INFL_PERC
```
